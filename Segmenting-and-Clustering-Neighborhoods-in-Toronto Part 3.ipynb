```python
import pandas as pd

import requests
from bs4 import BeautifulSoup
from scrapy.selector import Selector
import matplotlib.cm as cm
import matplotlib.colors as colors

from sklearn.cluster import KMeans


import folium
```

```python
url="https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
```

<h3>Brings HTML code from the given URL </h3>

```python
res=requests.get(url)
```

```python
res.text
```

<h3>Beautifies the HTML code to understand it.</h3>

```python
soup=BeautifulSoup(res.text)
soup
```

```python
response= Selector(text=res.text)
```

<h3>Extracts the data in tabular form on the webiste</h3>

```python
a=response.xpath('//table[@class = "wikitable sortable"]//text()').extract()
a
```

```python
col=["Postal Code","Borough","Neighbourhood"]
```

```python
df=pd.DataFrame(columns=col)
df
```

<h3>Converts HTML into Datafram</h3>

```python
for x in range(len(a)):
    if a[x]=='\n' and a[x+1]=='\n':
        postal=a[x-5]
        neigh=a[x-1]
        boro=a[x-3]
        postal=postal.replace("\n","")
        neigh=neigh.replace('\n',"")
        boro=boro.replace('\n',"")
        if boro !='Not assigned':

            df.loc[x+1]=[postal,boro,neigh]
```

```python
df=df[df['Postal Code'] !='Postal Code']
df.reset_index(inplace=True,drop=True)
```

```python
data=pd.read_csv("Coordinates.csv")
```

```python
df1=df
raw=data
```

```python
merged = pd.merge(df1, data, on='Postal Code')

merged
```

```python
toro=merged[merged['Borough'].str.contains("Toronto")]
toro
```

```python
map_toro = folium.Map(location=[43.6532, -79.3832], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(toro['Latitude'], toro['Longitude'], toro['Borough'], toro['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toro)  
    
map_toro
```
